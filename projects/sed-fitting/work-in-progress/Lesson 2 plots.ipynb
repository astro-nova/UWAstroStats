```python
rng = np.random.default_rng(seed=0)
```

```python
N = 21
xs = np.linspace(-np.pi, np.pi, N)
ys = np.sin(xs) + rng.normal(loc=0, scale=0.1, size=N)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(xs, ys, color='k')

xs_new = np.linspace(-3.6, 3.6, 101)
plt.plot(xs_new, np.sin(xs_new), ':', color='C3')
plt.ylim(-1.39, 1.39)
plt.xlim(-3.5,3.5)
```

```python
xdeg = np.arange(0, 15)
xarrs = []
for deg in xdeg:
    xarr = xs**deg
    xarrs.append(xarr)
xarrs = np.array(xarrs)
```

```python
fitter = LinearRegression(fit_intercept=False)
fitter = Lasso(fit_intercept=False, alpha=0.1, max_iter=5000)
fitter.fit(xarrs.T, ys)
```

```python
yfit=fitter.predict(xarrs.T)
```
